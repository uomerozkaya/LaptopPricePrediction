```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ilk_liste = []
x_detays = []
nmes = []

syf = 1

while syf<=500:
    r = requests.get("https://www.trendyol.com/laptop-x-c103108?pi=1"+str(syf)+"")
    soup = BeautifulSoup(r.content,"lxml")

    st1 = soup.find("div",attrs={"class":"prdct-cntnr-wrppr"})
    st2 = st1.find_all("div",attrs={"class":"p-card-wrppr"})

    for detaylar in st2:
        link_sonu = detaylar.a.get("href")
        link_basi = "https://www.trendyol.com"
        link = link_basi+link_sonu
        print(link)
        r1 = requests.get(link)
        soup1 = BeautifulSoup(r1.content,"lxml")

        fiyat = soup1.find("span",attrs={"class":"prc-slg"}).text
        print(fiyat)

        urun_adi = soup1.find("h1",attrs={"class":"pr-new-br"}).text
        print(urun_adi)

        #urun_markasi = soup1.find("a",attrs={"href":"/monster-x-b105536"})
        #print(urun_markasi)

        urun_detaylar = soup1.find_all("li",attrs={"class":"detail-attr-item"})
        
        t = []
        x_detay = []
        nme = []
        for detay in urun_detaylar:
            detay_a = detay.find("b").text
            b = detay.find("span").text
            #print(detay.find("b").text)
            x_detay.append(detay_a)
            nme.append(b)
            #dict1 = {tuple(nme) : tuple(x_detay)}
            
        
        #df2 = pd.DataFrame([x_detay],columns = nme)
        ilk_liste.append(fiyat)     
        x_detays.append([x_detay])
        nmes.append([nme])
        
        

    syf = syf + 1   
```

```python
#dict2 = {nme : x_detay}
for i in nme:
    i.remove("'")
    print(i)
```

```python
fiyats = []
for f in ilk_liste:
    removed = f.strip('TL ')
    fiyats.append(removed)
```

```python
detay_son = []
nme_son = []
for i in x_detays:
    detay_son.append(str(i).split(', '))
for j in nmes:
    nme_son.append(str(j).split(', '))
    
dict_list = []
for i,j in zip(nme_son,detay_son):
    dict1 ={n:d for n,d in zip(i,j)}
    dict_list.append(dict1)

df1 = pd.DataFrame(dict_list)
df1['Ürün Fiyatı'] =  fiyats
df1
```
